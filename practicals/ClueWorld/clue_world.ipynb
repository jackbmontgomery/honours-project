# Imports

```python
import os
import sys
import pathlib
import numpy as np

from pymdp.agent import Agent
from pymdp import utils, maths

import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.cm as cm
```

# World Definition

```python
grid_dims       = [6, 6] # dimensions of the grid (number of rows, number of columns)
num_grid_points = np.prod(grid_dims) # total number of grid locations (rows X columns)
grid            = np.arange(num_grid_points).reshape(grid_dims)

start_location  = (0, 0)
clue_location   = (2, 2)
reward_location = (3, 3)
```

```python
fig, ax = plt.subplots()
ax.imshow(grid, cmap='Blues', alpha=0.1)  # Light background for grid

# Plotting the clue and reward locations
ax.scatter(*start_location[::-1], color='green', s=200, label='Reward Location')
ax.scatter(*clue_location[::-1], color='blue', s=200, label='Clue Location')
ax.scatter(*reward_location[::-1], color='red', s=200, label='Reward Location')

# Adding grid lines
ax.set_xticks(np.arange(-0.5, grid_dims[1], 1), minor=True)
ax.set_yticks(np.arange(-0.5, grid_dims[0], 1), minor=True)
ax.grid(which='minor', color='black', linestyle='-', linewidth=2)

# Adding labels and title
ax.set_xticks(np.arange(grid_dims[1]))
ax.set_yticks(np.arange(grid_dims[0]))
ax.set_xticklabels(np.arange(1, grid_dims[1] + 1))
ax.set_yticklabels(np.arange(1, grid_dims[0] + 1))
ax.set_title('Grid with Clue and Reward Locations')
ax.legend()

plt.show()
```
