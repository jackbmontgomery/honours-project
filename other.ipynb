```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
plt.style.use('ggplot')
```

```python
# Set up data
size = 100
blanket_states = np.linspace(0, 1, size)
mu = -1 * blanket_states + np.random.normal(0, 1/4, size)
x = 2 * blanket_states + np.random.normal(0, 1/4, size)

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Plot the data
axs[0].scatter(blanket_states, mu, color='b', edgecolor='k', alpha=0.7)
axs[0].set_title(r'$\mu | b$')
axs[0].set_xlabel(r'$b$')
axs[0].set_ylabel(r'$\mu$')

axs[1].scatter(blanket_states, x, color='g', edgecolor='k', alpha=0.7)
axs[1].set_title(r'$x | b$')
axs[1].set_xlabel(r'$b$')
axs[1].set_ylabel(r'$x$')

# Scatter plot for x and mu
axs[2].scatter(x, mu, color='r', edgecolor='k', alpha=0.7)
axs[2].set_title(r'$\mu$ vs $x$')
axs[2].set_xlabel(r'$x$')
axs[2].set_ylabel(r'$\mu$')

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(x, mu)

# Generate line of best fit
line_x = np.linspace(min(x), max(x), 100)
line_y = slope * line_x + intercept

# Plot line of best fit
axs[2].plot(line_x, line_y, color='black', linestyle='--', linewidth=2, label=f'Best fit')
axs[2].legend()

# Add padding between plots
plt.tight_layout()

# Show the plot
plt.show()
```
