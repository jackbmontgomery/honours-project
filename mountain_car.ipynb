```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')
```

```python
results_directoy    = './message_passing/ai_simulation_data.csv'
data = np.loadtxt(results_directoy, delimiter=',', skiprows=1)

time            = data[:, 0]
actions         = 0.04 * np.tanh(data[:, 1])
observations    = data[:, 2]
free_energy     = data[:, 3]
```

```python
fig = plt.figure(figsize=(12, 6))
gs = fig.add_gridspec(2, 2, height_ratios=[1, 1], width_ratios=[1, 1], hspace=0.3, wspace=0.2)

# Create subplots
ax1 = fig.add_subplot(gs[0, 0])  # Top left
ax2 = fig.add_subplot(gs[0, 1])  # Top right
ax3 = fig.add_subplot(gs[1, :])  # Bottom (spanning both columns)

# Plot data
ax1.plot(time, observations, color='orange')
ax1.set_ylabel("Position Observation")
ax1.set_xlabel("Time")

ax2.plot(time, free_energy, color='green')
ax2.set_ylabel("Free Energy")
ax2.set_xlabel("Time")

ax3.plot(time, actions, color='blue')
ax3.set_ylabel("Force Applied")
ax3.set_xlabel("Time")

# Set titles for clarity
ax1.set_title("Position Observation vs. Time")
ax2.set_title("Free Energy vs. Time")
ax3.set_title("Force Applied vs. Time")

# Show plot
plt.show()
```
