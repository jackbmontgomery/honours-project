```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')
```

```python
results_directoy    = './message_passing/ai_simulation_data.csv'
data = np.loadtxt(results_directoy, delimiter=',', skiprows=1)

time            = data[:, 0]
actions         = 0.04 * np.tanh(data[:, 1])
observations    = data[:, 2]
free_energy     = data[:, 3]
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

axs[0].plot(time, observations)
axs[0].set_ylabel("Position Observation")
axs[0].set_xlabel("Time")

axs[1].plot(time, actions)
axs[1].set_ylabel("Force Applied")
axs[1].set_xlabel("Time")

plt.show()
```
